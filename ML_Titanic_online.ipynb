```python
import pandas as pd
traindf = pd.read_csv("./data/titanic/train.csv", encoding="utf-8")
traindf
```

```python
testdf = pd.read_csv("./data/titanic/test.csv", encoding="utf-8")
testdf
```

```python
# isna(), 判斷每格是否為空值
# sum(), 將空值數量加總
traindf.isna().sum()
```

```python
# 1. 填補缺失值：有大小關係填中位數
med = traindf.median()
traindf = traindf.fillna(med)
traindf.isna().sum()
```

```python
# (測試資料填補) 千萬不要重新算，用訓練算出的來補
testdf = testdf.fillna(med)
testdf.isna().sum()
```

```python
# value_counts(), 統計出類別資料各幾個
# idxmax(), 找出個數最多的類別
most = traindf["Embarked"].value_counts().idxmax()
traindf["Embarked"] = traindf["Embarked"].fillna(most)
traindf.isna().sum()
```

```python
testdf["Embarked"] = testdf["Embarked"].fillna(most)
testdf.isna().sum()
```

```python
# 2. 對類別(?種)特徵 -> One-Hot Encoding(不能承認他們有大小關係)
# 轉0, 1, 2 會間接承認大小關係
dummy = pd.get_dummies(traindf["Embarked"])
# axis=0直的接，axis=1橫的接
traindf = pd.concat([traindf, dummy], axis=1)
traindf
```

```python
dummy = pd.get_dummies(testdf["Embarked"])
# axis=0直的接，axis=1橫的接
testdf = pd.concat([testdf, dummy], axis=1)
testdf
```

```python
dummy = pd.get_dummies(traindf["Sex"])
traindf = pd.concat([traindf, dummy], axis=1)
dummy = pd.get_dummies(testdf["Sex"])
testdf = pd.concat([testdf, dummy], axis=1)
traindf
```

```python
# 3. 處理比較困難的欄位
mid = "Braund, Mr. Owen Harris".split(",")[-1].split(".")[0]
mid = mid.replace(" ", "")
mid
```

```python
# pandas apply(流程名字)
# 如果遇到種類很多很多的時候，看一下每種有幾個
def nameflow(s):
    mid = s.split(",")[-1].split(".")[0]
    mid = mid.replace(" ", "")
    return mid
name = traindf["Name"].apply(nameflow)
pd.crosstab(name, traindf["Survived"]).T # 將欄位直的轉橫的
```

```python
def nameflow(s):
    mid = s.split(",")[-1].split(".")[0]
    mid = mid.replace(" ", "")
    if mid == "Miss":
        return "Miss"
    elif mid == "Mr":
        return "Mr"
    elif mid == "Mrs":
        return "Mrs"
    else:
        return "Throw"
name = traindf["Name"].apply(nameflow)
dummy = pd.get_dummies(name)
traindf = pd.concat([traindf, dummy], axis=1)
traindf
```

```python
name = testdf["Name"].apply(nameflow)
dummy = pd.get_dummies(name)
testdf = pd.concat([testdf, dummy], axis=1)
testdf
```

```python
# 要丟掉東西，印一下看欄位是哪些
traindf.columns
```

```python
# traindf先把PassenferId Drop, test先留著
traindf = traindf.drop(['PassengerId', 'Name', 'Sex', 'Ticket',
                        'Cabin', 'Embarked', 'Throw'], axis=1)
testdf = testdf.drop(['Name', 'Sex', 'Ticket', 'Cabin',
                      'Embarked', 'Throw'], axis=1)
traindf
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
# numpy：處理大量數字
import numpy as np

trainx = traindf.drop(["Survived"], axis=1)
trainy = traindf["Survived"]
# n_estimators -> 要有幾棵樹
clf = RandomForestClassifier(max_depth=7, n_estimators=28)
np.average(cross_val_score(clf, trainx, trainy, cv=10))
# cv=10：切10份
```

```python
from sklearn.model_selection import GridSearchCV

# range(5, 11) = [5, 6, 7, 8, 9 ,10]
p = {
    "max_depth":range(5, 11),
    "n_estimators":range(20, 30)
}
clf = RandomForestClassifier()
s = GridSearchCV(clf, p, cv=10)
s.fit(trainx, trainy)
print(s.best_params_)
print(s.best_score_)
```

```python
testx = testdf.drop(["PassengerId"], axis=1)
testid = testdf["PassengerId"]
clf = RandomForestClassifier(max_depth=7, n_estimators=27)
clf.fit(trainx, trainy)
pre = clf.predict(testx)
result = pd.DataFrame()
result["PassengerId"] = testid
result["Survived"] = pre
# index=False：不要儲存列編號(0, 1, 2...)
result.to_csv("result.csv", encoding="utf-8", index=False)
result
```
