```python
from sklearn.datasets import load_boston
import pandas as pd

boston = load_boston()
df = pd.DataFrame(boston["data"], columns=boston["feature_names"])
df["target"] = boston["target"]
df
```

```python
# (有看過的資料)作出公式 -> (沒看過的資料)驗證公式好壞
from sklearn.model_selection import train_test_split

# train_test_split -> (特徵90%, 特徵10%, 目標90%, 目標10%)
x_train, x_test, y_train, y_test = train_test_split(df.drop(["target"], axis=1), df["target"], test_size=0.1)
# axis=0是橫列 axis=1是直行
```

```python
# 訓練
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor(max_depth=5)
reg.fit(x_train, y_train)
```

```python
from sklearn.tree import export_graphviz
import graphviz

g = export_graphviz(reg, out_file=None,
                    feature_names=boston["feature_names"],
                    filled=True)
# out_file=None：不要輸出檔案
graph = graphviz.Source(g)
graph
```

```python
# 驗證公式好壞
# 預測(predict)
print("預測結果：", list(reg.predict(x_test)))
print("真正結果：", list(y_test))

from sklearn.metrics import r2_score
print("r2 score：", r2_score(y_test, reg.predict(x_test)))
```
