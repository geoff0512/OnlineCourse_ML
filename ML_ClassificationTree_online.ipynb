```python
from sklearn.datasets import load_iris
iris = load_iris()
iris
```

```python
import pandas as pd
df = pd.DataFrame(iris["data"], columns=iris["feature_names"])
df["target"] = iris["target"]
# df.to_csv("iris.csv", encoding="utf-8", index=False)
df
```

```python
# 視覺化 相關係數圖
# seaborn是存在於matplotlib裡的套件
# %matplotlib inline 代表只要有圖就show出來 不用再print
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

plt.figure(figsize=(14, 10)) # 變更圖的大小
sns.heatmap(df.astype("float").corr(), cmap="YlGn", annot=True)
# astype("float") 確認裡面都是小數，整數進corr會有問題
# annot=True 將相關係數打上去
# cmap：更改圖的顏色
```

```python
df.corr()
```

```python
from sklearn.model_selection import train_test_split

# train_test_split -> (特徵90%, 特徵10%, 目標90%, 目標10%)
x_train, x_test, y_train, y_test = train_test_split(df.drop(["target"], axis=1), df["target"], test_size=0.1)
# axis=0是橫列 axis=1是直行
```

```python
# 訓練
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=3) # 先剪枝調整max_depth
clf.fit(x_train, y_train)
```

```python
from sklearn.tree import export_graphviz
import graphviz

g = export_graphviz(clf, out_file=None, feature_names=iris["feature_names"],
                    class_names=iris["target_names"], filled=True, special_characters=True)
# out_file=None：不要輸出檔案
graph = graphviz.Source(g)
graph
```

```python
pre = clf.predict(x_test)
print("預測結果：", list(pre))
print("真正結果：", list(y_test))
```

```python
from sklearn.metrics import accuracy_score
print("預測成功機率：", accuracy_score(pre, y_test))
```

```python
from sklearn.metrics import confusion_matrix
pd.DataFrame(confusion_matrix(y_test, pre))
# 不在斜線軸上的是錯誤的
```
